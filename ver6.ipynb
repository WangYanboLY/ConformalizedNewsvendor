# jackknife+ modification
用LOO方法构造预测和修正项

```python
import numpy as np
import ConformaQuantile as CQ

quantile = 0.8
n_samples = 3000
n_X1 = 20
n_X2 = 2
n_X3 = 2
np.random.seed(0)

X1 = abs(np.random.normal(6.4, 10, (n_samples, n_X1)))
X2 = abs(np.random.normal(0.4, 1, (n_samples, n_X2)))
X3 = abs(np.random.normal(0.9, 1, (n_samples, n_X3)))

coefficients = abs(np.random.normal(10, 400, n_X1 + n_X2))
X = np.hstack((X1, X2, X3))
noise = np.random.normal(0, 1, n_samples)

X_true = X[:, :(n_X1 + n_X2)]
X_observed = np.hstack((X1, X3))
Y = np.dot(X_true, coefficients)


train_ratio = 0.6
validation_ratio = 0.2
test_ratio = 0.2
```

```python
loss = {}
models = ['linear', 'lasso', 'ridge', 'random_forest', 'glm', 'neural_network']

Y0 = Y + noise
for model in models:
    CQ.perform_jackknife_plus(X_observed, Y0, validation_ratio, quantile, model)
```
